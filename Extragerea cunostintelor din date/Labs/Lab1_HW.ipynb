```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load data from csv
data = pd.read_csv("loan.csv")

# Get number of instances and attributes
data.shape # -> 614 instances and 13 attributes
```

```python
# First 5 instances
data.head()
```

```python
# Data type for each attribute
data.dtypes
```

```python
# Summary for every numerical attribute in the data
data.describe()
```

```python
# Summary for every categorical attribute in the data
data.describe(include=[object])
```

```python
# Instances distribution on 'Loan_Status' class
class_distribution = data['Loan_Status'].value_counts()
class_distribution
```

```python
plt.figure(figsize=(8, 6))
class_distribution.plot(kind='bar', color=['pink', 'skyblue'])
plt.title('Distribution of Instances in Classes')
plt.xlabel('Loan Status')
plt.ylabel('Number of Instances')
plt.xticks(rotation=0)
plt.show()
```

```python
# Instances distribution on other relevant classes
for column in ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Loan_Amount_Term', 'Credit_History', 'Property_Area']:

    print("Column:", column)
    print(data[column].value_counts())
    print("\n")
```
