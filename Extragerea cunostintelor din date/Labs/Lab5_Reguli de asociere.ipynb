### Example 1:  Package apyori - https://pypi.org/project/apyori/

#### Remarks - apyori / Observații - apyori:
 * each RelationRecord corresponds to a frequent itemset / structura RelationRecord corespunde unui set de entități
 * each OrderedStatistics correspond to a rule (the antecedent is specified by items_base, the consequent is specified by items_add)/ fiecare componenta de tip OrderedStatistic corespunde unei reguli (antecedent: items_base, concluzie: items_add)

```python
!pip install apyori
import apyori
from apyori import apriori
```

```python
# first example - small set of transactions
transactions = [['milk', 'bread', 'meat', 'water'],
['bread', 'water'],
['bread', 'butter', 'meat', 'water'],
['water'],
]
results = list(apriori(transactions))
```

#### Example 2 - extract association rules from the groceries list of transactions / Exemplu 2 - extragere reguli de asociere din fisier groceries.csv
Hint:  https://www.kaggle.com/code/raibabar/market-basket-analysis-with-apriori-algorithim

```python
transactions_groceries = []

with open('groceries.csv') as f:
    for line in f:
        transaction = [item for item in line.strip().split(',') if item != 'NaN']
        transactions_groceries.append(transaction)
```

```python
len(transactions_groceries)
```

```python
# extragere reguli folosind algoritmul apriori
rules_groceries = list(apriori(
    transactions_groceries, 
    min_support=0.001, 
    min_confidence=0.90,
    min_length=2,
    max_length=5, min_lift=4))
```

```python
len(rules_groceries)
```

```python
# construct a data frame containing the rules / construirea unui tabel cu regulile
rules_df = pd.DataFrame(
    [{'From': list(rule[0])[0],
    'To': list(rule[0])[1],
    'Support': rule[1],
    'Confidence': rule[2][0][2],
    'Lift': rule[2][0][3]} for rule in rules_groceries])

# sort the rules decreasingly by lift / sortare descrescatoare dupa lift
rules_df = rules_df.sort_values('Lift', ascending=False)
```

### Example 3.  FPgrowth algorithm (https://github.com/chonyy/fpgrowth_py) /  Algoritmul FPgrowth  (https://github.com/chonyy/fpgrowth_py)

```python
!pip install fpgrowth_py
```

```python
from fpgrowth_py import fpgrowth
transactions = [['milk', 'bread', 'meat', 'water'],
['bread', 'water'],
['bread', 'butter', 'meat', 'water'],
['water'],
]
freqItemSet, rules = fpgrowth(transactions, minSupRatio=0.3, minConf=0.75)
print("Frequent itemsets:",freqItemSet)
print("Rules (antecedent, consequent, confidence):",rules)  
```

### Example 4. Package mlxtend (Machine Learning Library Extensions) - apriori algorithm 
Source: https://github.com/jsrpy/Association_Rule/blob/master/association_rules_introduction.ipynb

```python
!pip install mlxtend
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# data in one-hot encoded format
data = {'ID':[1,2,3,4,5,6],
       'Onion':[1,0,0,1,1,1],
       'Potato':[1,1,0,1,1,1],
       'Burger':[1,1,0,0,1,1],
       'Milk':[0,1,1,1,0,1],
       'Beer':[0,0,1,0,1,0]}
df = pd.DataFrame(data)
df
```

```python
frequent_itemsets = apriori(df[['Onion', 'Potato', 'Burger', 'Milk', 'Beer' ]], min_support=0.50, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules
```

```python
# select rules based on confidence and lift
rules [ (rules['lift'] >1.125)  & (rules['confidence']> 0.8)  ]
```
